# 处理网页文本

```python
import pandas as pd
import re
from datetime import datetime, timedelta

city = '西安'

# 从网页获取url文件
# url复制到八爪鱼，爬取获得全文文本
df = pd.read_excel(rf"H:\微信公众平台.xlsx")

# 可选
df = df.drop(columns=['标题'])
# 移除空格和换行符
df['文本'] = df['文本'].astype(str).str.replace(r'\n', '', regex=True)
df['文本'] = df['文本'].astype(str).str.replace(r'\s+', '', regex=True)

# 添加日期列，排序
def match_date(row):
    date_match = re.search(r'(\d+)月(\d+)日0-24时', row)
    if date_match:
        month = int(date_match.group(1))
        day = int(date_match.group(2))
        year = 2022
        date = datetime(year, month, day).strftime('%Y-%m-%d')
        return date
df['date'] = df['文本'].apply(match_date)
df = df.sort_values(by=['date'])
df['date'] = pd.to_datetime(df['date'])
print(df.head())


# 合并从网页获取的url
df_url = pd.read_excel(rf"H:\1_{city}链接.xlsx") #url所在

# 如果标题不带时间，要用‘发表时间'改。由于发表时间是第二天，所以减一天
df_url['date'] = pd.to_datetime(df_url['发表时间'].str.extract(r'(\d{4}-\d{2}-\d{2})')[0])- pd.Timedelta(days=1)

# 如果标题带有时间，用这个
# df_url['date'] = df_url['标题'].apply(match_date) #获得疫情对应时间

df_url = df_url.sort_values(by=['date'])
df_url = df_url[df_url['date'] >= '2022-06-01']
df_url = df_url.drop(columns=['标题','发表时间'])  #移除不需要的列
print(df_url.head())

df = df.merge(df_url, on=['date'], how='right')
print(df)
df = df.drop_duplicates()
print(df.columns)
df.to_csv(rf"H:\2_{city}网页文本.csv",encoding='utf-8-sig',index=False)
```

# 第一版本正则表达式

```python

import pandas as pd
from datetime import datetime, timedelta
import re


# 北京区县字典准备
regions = [
    "东城区", "西城区", "朝阳区", "丰台区", "石景山区", "海淀区", "门头沟区",
    "房山区", "通州区", "顺义区", "昌平区", "大兴区", "怀柔区", "平谷区", "密云区", "延庆区", "经开区"
]
# 初始化字典
result_dict = {}


# 读取 CSV 文件
df = pd.read_csv(rf"H:\2_北京网页文本.csv")
for index, row in df.iterrows():
    # 提取日期
    date_match = re.search(r'(\d+)月(\d+)日', row['标题'])
    if date_match:
        month = int(date_match.group(1))
        day = int(date_match.group(2))
        year = 2022
        date = datetime(year, month, day).strftime('%Y-%m-%d')


        # 如果该日期在 result_dict 中尚未存在，则初始化该日期的字典
        if date not in result_dict:
            result_dict[date] = {region: 0 for region in regions}

    # 提起疫情数据
    for region in regions:
        # 初始化计数
        total_cases = 0

        # 匹配规则 1：匹配逗号分隔的多个编号
        pattern1 = rf'确诊病例(\d+)、(\d+)：现住{region}'
        matches1 = re.findall(pattern1, row['真正内容'])
        print(matches1)
        for start, end in matches1:
            total_cases += int(end) - int(start) + 1

        # 匹配规则 2：匹配编号范围（如 1至2）
        pattern2 = rf'确诊病例(\d+)至(\d+)：现住{region}'
        matches2 = re.findall(pattern2, row['真正内容'])
        for start, end in matches2:
            total_cases += int(end) - int(start) + 1

        # 匹配规则 3：匹配单个编号（如 1）
        pattern3 = rf'确诊病例(\d+)：现住{region}'
        matches3 = re.findall(pattern3, row['真正内容'])
        total_cases += len(matches3)

        # 匹配规则 4：没有编号的情况
        pattern4 = rf'确诊病例：现住{region}'
        matches4 = re.findall(pattern4, row['真正内容'])
        total_cases += len(matches4)  # 每次匹配到此情况，计数加 1

        result_dict[date][region] = total_cases


# 创建一个空的列表，用于存储每一行数据
data = []

# 遍历嵌套字典，将数据转换为列表的形式
for date, regions in result_dict.items():
    for region, value in regions.items():
        data.append([date, region, value])

# 将列表转换为 DataFrame
total_df = pd.DataFrame(data, columns=['date', 'region', 'value'])

# 输出 DataFrame
print(total_df)
# total_df.to_csv(rf"H:\3_北京结果.csv",encoding='utf-8-sig',index=False)






```

# 正则表达式-适用北京

```python
import re
from collections import Counter

# 示例文本
text = """
确诊病例62至65、146、147：现住丰台区；
确诊病例60、61、154、155：现住丰台区；
确诊病例54、55、56、151、152、153：现住通州区；
确诊病例8至50、52、53、80至117：现住朝阳区;
确诊病例62至65：现住坪山区；
确诊病例88、89：现住坪山区；
确诊病例103：现住坪山区；
确诊病例：现住坪山区；
"""

# 北京区县字典准备
regions = [
    "东城区", "西城区", "朝阳区", "丰台区", "石景山区", "海淀区", "门头沟区",
    "房山区", "通州区", "顺义区", "昌平区", "大兴区", "怀柔区", "平谷区", "密云区", "延庆区", "经开区"
]
result_dict = {}

df = pd.read_csv(rf"H:\消费相关数据\区县层面疫情\2_北京网页文本.csv")
for index, row in df.iterrows():
    # 提取日期
    date_match = re.search(r'(\d+)月(\d+)日', row['标题'])
    if date_match:
        month = int(date_match.group(1))
        day = int(date_match.group(2))
        year = 2022
        date = datetime(year, month, day).strftime('%Y-%m-%d')


        # 如果该日期在 result_dict 中尚未存在，则初始化该日期的字典
        if date not in result_dict:
            result_dict[date] = {region: 0 for region in regions}

    # 提起疫情数据
    for region in regions:
        # 初始化计数
        pattern = rf"确诊病例([\d、至]*)：现住{region}"
        matches = re.findall(pattern, row['真正内容'])

        count = 0
        for case_range in matches:
            if case_range:  # 如果有病例编号
                cases = re.findall(r"(\d+)(?:至(\d+))?", case_range)
                for start, end in cases:
                    if end:  # 如果存在范围
                        count += int(end) - int(start) + 1
                    else:  # 单个病例
                        count += 1
            else:  # 如果没有病例编号
                count = 1

            result_dict[date][region] = count


# 遍历嵌套字典，将数据转换为df的形式
data = []
for date, regions in result_dict.items():
    for region, value in regions.items():
        data.append([date, region, value])
total_df = pd.DataFrame(data, columns=['date', 'region', 'value'])


print(total_df)
total_df.to_csv(rf"H:\消费相关数据\区县层面疫情\3_北京结果2.csv",encoding='utf-8-sig',index=False)

```

```python
import pandas as pd
import numpy as np

cities = {
    "北京": "Beijing",
    "上海": "Shanghai",
    "广州": "Guangzhou",
    "深圳": "Shenzhen",
    "成都": "Chengdu",
    "重庆": "Chongqing",
    "武汉": "Wuhan"
}
# 定义日期范围
date_range = pd.date_range(start='2022-06-01', end='2022-12-21') # DatetimeIndex


df = pd.DataFrame()
for city,pinyin in cities.items():
    merged_df = pd.read_csv(rf"H:\\3_{city}结果.csv")
    merged_df['city'] = pinyin

    # check data
    merged_df['date'] = pd.to_datetime(merged_df['date']).dt.strftime('%Y-%m-%d')
    city_dates = pd.to_datetime(merged_df['date'].unique()) # convert to DatetimeIndex
    missing_dates = {}
    missing = date_range.difference(city_dates)
    if not missing.empty:
        missing_dates[city] = missing.strftime('%Y-%m-%d').tolist()

    # add missing data to df
    if missing_dates:
        new_rows = []
        for date in missing_dates[city]:
            for region in merged_df['region'].unique():
                new_rows.append({'date': date, 'region': region, 'value': np.nan, 'city': pinyin})
        missing_df = pd.DataFrame(new_rows)
        print(missing_df)
        merged_df = pd.concat([merged_df, missing_df], ignore_index=True)
    else:
        print(rf"{city}数据齐全")

    # 按日期和区域排序
    # merged_df['date'] = pd.to_datetime(merged_df['date'])
    merged_df = merged_df.sort_values(by=['date', 'region']).reset_index(drop=True)
    merged_df.to_csv(rf"H:\3_{city}结果.csv", encoding='utf-8-sig', index=False)

    # merge df
    if df.empty:
        df = merged_df
    else:
        df = pd.concat([df, merged_df], ignore_index=True)

# 打印缺失的日期
for city, dates in missing_dates.items():
    print(f"{city} 缺失的日期: {dates}")


df.to_csv(rf"H:\\确诊.csv",encoding='utf-8-sig',index=False)
```
